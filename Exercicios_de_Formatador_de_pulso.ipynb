```python
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt
import random
```

# Recriando o código do livro

```python
#Gerando 1000 simbolos 4 PAM
N = 1000
M = 10 #largura de faixa inicial do pulso
simbolos = np.zeros(N)
for i in range(N):
  simbolos[i] = random.choice([-3,-1,1,3])
```

```python
sim_up = np.zeros(N*M)
sim_up[::M] = simbolos #Oversample dos simbolos

pulso = signal.windows.hamming(M) #Pulso com largura de faixa T

out = np.convolve(pulso, sim_up) #Convoluindo o pulso e os simbolos

indices_t = np.arange(4, len(out), M)

figure, axis = plt.subplots(1,2, figsize=(18,5))

# Plotando a saida
axis[0].plot(out)
axis[0].set_xlim([0,250])
axis[0].scatter(indices_t, out[indices_t], color='red', label='Pontos Destacados')
axis[0].set_title('saida do formatador de pulso')
axis[0].set_xlabel('simbolos')
axis[0].grid(True)

# Calculando a resposta em frequência da saida
fft_out = np.fft.fft(out)

# Calculando as frequências correspondentes
freqs = np.fft.fftfreq(len(fft_out))/0.5

# Plotando o espectro de frequência do pulso
axis[1].semilogy(freqs, abs(fft_out))
axis[1].set_title('Espectro da saida do formatador de pulso')
axis[1].set_xlabel('Frequência')
axis[1].set_ylabel('Magnitude')
axis[1].set_xlim([0,1])
axis[1].set_ylim([1e-1,1e4])
axis[1].grid(True)

plt.show()
```

# 11.14

Plotando o cosseno levantado para cada valor de roll off

```python
def raised_cosine(t, roll_off):
    if roll_off == 0:
        return np.sinc(t/np.pi)
    else:
        denom = np.sqrt(np.pi) * np.sqrt(2) * roll_off
        num = np.sin(np.pi * t * (1 - roll_off)) + 4 * roll_off * t * np.cos(np.pi * t * (1 + roll_off))
        return num / (denom * t * (1 - (4 * roll_off * t) ** 2))
```

```python
# Parâmetros
t = np.linspace(-10, 10, 1000)  # Tempo
roll_off_values = [0.1, 0.3, 0.5, 0.9]  # Valores do fator de roll-off
sampling_freq = 1000  # Frequência de amostragem

# Plot
plt.figure(figsize=(14, 8))

# Plot do domínio do tempo
plt.subplot(2, 1, 1)
for roll_off in roll_off_values:
    plt.plot(t, raised_cosine(t, roll_off), label=f'Roll-off = {roll_off}')
plt.title('Cosseno Levantado para Diferentes Valores de Roll-off (Domínio do Tempo)')
plt.xlabel('Tempo')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

# Plot do domínio da frequência
plt.subplot(2, 1, 2)
for roll_off in roll_off_values:
    sinal = raised_cosine(t, roll_off)
    spectrum = np.fft.fftshift(np.fft.fft(sinal))
    freq = np.fft.fftshift(np.fft.fftfreq(len(sinal), d=1/sampling_freq))
    norm_spectrum = np.abs(spectrum) / np.max(np.abs(spectrum))
    norm_freq = freq/max(freq)
    plt.plot(norm_freq, norm_spectrum, label=f'Roll-off = {roll_off}')

plt.title('Espectro de Frequência Normalizado de Cosseno Levantado para Diferentes Valores de Roll-off')
plt.xlabel('Frequência (Hz)')
plt.ylabel('Magnitude Normalizada')
plt.xlim(-0.2, 0.2)
plt.ylim(0, 1.2)
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
